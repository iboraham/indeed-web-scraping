```python
df=pd.read_csv('jobs data/data_scientist.csv')
len(df)
```

```python
def empty_count(str_series):
    counter = 0
    for ind in str_series:
        if ind == []:
            counter+=1
    return counter
```

```python
desc=df.Description.str.lower()
qualifs=desc.str.findall(r'((?:qualifications|requirements|desirable|skills|required|role)+(?:\:|\?|\n)(?:.|\s)*)')
qualifs[25],empty_count(qualifs)
```
